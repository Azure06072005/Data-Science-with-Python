### NumPy Array
You can plot NumPy arrays using Matplotlib using the same syntax as what we saw earlier with Python lists.
Note that by default, Matplotlib connects consecutive points with a line when drawing a vanilla 2D plot (using the plot function). Since were looking to view the vectors in isolation (and not by connecting them together with a line), we utilize a scatter plot (using the scatter function) which helps observe data points individually.

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([1,2])
y = np.array([3,4])
plt.scatter(x, y, color = ["r", "b"])
plt.show()
```

### 2D Vectors
To plot 2D vectors using Matplotlib, we use Pythons inbuilt zip() function which aggregates elements from its inputs by returning an iterator of tuples where the ith tuple contains the ith element from each of the input iterables. To read more about the zip() function, refer our Python tutorial.
Using zip(), we populate lists with the X and Y components of the vectors u
and v

```python
import numpy as np 
import matplotlib.pyplot as plt 

a = np.array([4,5])
b = np.array([5,8])
x__coords, y__coords = zip(a,b)
plt.scatter(x__coords, y__coords, color = ["r", "b"])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def plot_vector2d(vector2d, origin = [0,0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1], head_width=0.05, head_length=0.2, length_includes_head=True, **options)
u = np.array([1,2])
v = np.array([3,4])

x, y = zip(u,v)

plt.scatter(x,y,color=["r", "b"])
plot_vector2d(u, color="r")
plot_vector2d(v, color="b")
plt.axis([0,9,0,6])
plt.show()
```

## Example 
Now, lets plot a mathematical function. We use NumPys linspace function (to learn more about NumPy and its linspace function, refer to our NumPy tutorial) to create an array x containing 500 floats ranging from 2 to 2, then we create a second array y computed as the square of x.

```python
import matplotlib.pyplot as plt  # Th vin v biu 
import numpy as np  # Th vin tnh ton s hc

x = np.linspace(-2, 2, 500)  # To mng x vi 500 gi tr t -2 n 2
y = x**2  # Tnh y = x^2 cho mi gi tr x

plt.plot(x, y)  # V  th ca y = x^2
plt.show()  # Hin th  th
```

you can get a sine wave: 

```python
# Import th vin matplotlib  v  th v numpy  thao tc vi mng s hc
import matplotlib.pyplot as plt  # Th vin v  th 
import numpy as np  # Th vin tnh ton mng s hc

# To mng cc gi tr x t -2 n 2 vi bc nhy 0.1
x = np.arange(-2*np.pi, 2*np.pi, 0.05)  # x = [-2, -2 + 0.1, -2 + 0.2, ..., 2 - 0.1]

# Tnh y = sin(x) cho mi gi tr ca x
y = np.sin(x)  # Gi tr y tng ng vi y = sin(x) ti mi x trong mng

# V  th ca y = sin(x) vi x trn trc honh v y trn trc tung
plt.plot(x, y)  # V  th ng cong ca hm y = sin(x)

# Hin th biu  trn mn hnh
plt.show()  # Hin th biu 
```

### Plotting 3D data
Plotting 3D data is relatively straightforward using Matplotlibs Axes3D. Similar to plotting 2D data, you can use Python lists or NumPy arrays as inputs to your plotting function, say plot or scatter.
Lets consider plotting 3D vectors as an example:

```python
import matplotlib.pyplot as plt 
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D

# 3D vectors
a = np.array([1,2,3])
b = np.array([4,5,6])

subplot3d = plt.subplot(111, projection = '3d')
x, y, z = zip(a, b)
subplot3d.scatter(x, y, z)
subplot3d.set_zlim3d([0, 10])
plt.show()
```

### Axes Extents
The axes in Matplotlib plots automatically match the extent of the data
if we wish to override the asex extents of plots, for e.g., to give the grapl a bit more room, we can call the axis function to change the extent of each axis with [xmin, xmax, ymin, ymax]

```python
import matplotlib.pyplot as plt 

plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7])
plt.show()
```

### Additional Plot Attributes
With just a little bit of extra work we can superimpose multiple lines on the same graph, add a title, legend, axis labels and draw a grid:

```python
import matplotlib.pyplot as plt
import numpy as np 

# Compute the X and Y coordinates for points on sine and cosine curves
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points
plt.plot(x, y_sin)
plt.plot(x, y_cos)

# Axes labels
plt.xlabel('Radians')
plt.ylabel('Magnitude')

# Custom X and Y axis limits
plt.xlim(np.ceil(-2*np.pi), np.floor(2*np.pi))
plt.ylim(-1, 1)

# Add a plot tittle 
plt.title('Sine and Cosine')

# Add a legend 
plt.legend(['Sine', 'Cosine'])

# Show the grid
plt.grid()
plt.show()
```
