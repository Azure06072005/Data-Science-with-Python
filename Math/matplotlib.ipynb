## Plotting 2D data
### Python list
the most important function in the matplotlib.pyplot module is plot, which allows you to plot 2D data.

```python
import matplotlib.pyplot as plt

plt.plot([1,2,4,8,16,8,4,2,1])
plt.show()                          # You must call plt.show() to render your final plot.
```

```python
plt.plot([-32,-16,-8,-4,-2,-1,0,1,2,4,8,16,32])
plt.show()
```

```python
plt.plot([-1,0,1,-1,0,1,2,-4,1,-2])
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot([-10,-40,-5,30,0], [5,10,20,40,30])
plt.show()  
```
